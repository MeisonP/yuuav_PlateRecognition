```python
#decoding: utf-8
#encoding: utf-8

#openCV transform C++ code to python
#python2.7  openCV3.4

import numpy as np
import cv2 as cv
import os
import logging
import time 
import shutil
#import argparse


def ini():  
    '''ap = argparse.ArgumentParser()
    ap.add_argument("-n", "--image_name", required=True,
                help="image name that use to process")
    ap.add_argument("-m", "--model_type", required=True,
                help="model type 0:normal,1:raw_blue,2:raw_yellow,  other value :raw_sobel")
    
    args = vars(ap.parse_args())'''
    
    
    TM=time.strftime("%Y-%m-%d %H-%M-%S",time.localtime())
    LOG_FORMAT = "%(levelname)s -[:%(lineno)d]- %(message)s"
    logging.basicConfig(level=logging.INFO, format=LOG_FORMAT)
    logging.info('**************************Mason(%s)*******************'%(TM,))
    
def show(image,name):
    cv.imshow('%s'%name,image)
    cv.waitKey(0)
    logging.info('show image....')
    
def save_image(image,name,path):
    path_out=path+name+'.jpg'
    cv.imwrite(path_out,image)
    logging.info('save image:%s ,done'%name)
    
def clean_folder(path):
    shutil.rmtree(path)  
    os.mkdir(path)
    logging.info('cleaning folder:%s done'%path)
    

#########################Color_seg func
#将RGB模型图像映射到HVS模型
def HSV(image,path):#image is a equliaed image
    logging.info('HSV...')
    image_HSV= cv.cvtColor(image, cv.COLOR_BGR2HSV)
    H, S, V = cv.split(image_HSV)
    save_image(H,'H',path)
    save_image(S,'S',path)
    save_image(V,'V',path)
    return image_HSV, H, S, V

def Gary(image):
    logging.info('Gary...')
    iamge_Gary=cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    return iamge_Gary

#直方图均衡/灰度均衡，使得输入图像在每一级灰度上都有相同的像素点输出,
#输入为单通道，所以对于多通道的RGB要3个通道单独输入
def Equalization(image_RGB,path):
    logging.info('RGB_Equalization...')
    B, G, R = cv.split(image_RGB)
    #print np.array(R).shape
    save_image(R,'R',path)
    save_image(G,'G',path)
    save_image(B,'B',path)
    
    R_Equal=cv.equalizeHist(R)
    save_image(R_Equal,'R_Equal',path)
    G_Equal=cv.equalizeHist(G)
    save_image(G_Equal,'G_Equal',path)
    B_Equal=cv.equalizeHist(B)
    save_image(B_Equal,'B_Equal',path)
    
    RGB_merged = cv.merge([B,G,R])
    image_Equal_RGB=RGB_merged
    return image_Equal_RGB


def Segment_Blue(image_orig,image_HSV,H,S,V):#must be HSV 
    logging.info('Segment_Blue...')
    H_Blue_Min=90
    H_Blue_Max=130
    S_Blue_Min=80
    S_Blue_Max=255
    V_Blue_Min=90
    V_Blue_Max=255
    LowerBlue = np.array([H_Blue_Min, S_Blue_Min,V_Blue_Min])
    UpperBlue = np.array([H_Blue_Max, S_Blue_Max,V_Blue_Max])
    mask= cv.inRange(image_HSV, LowerBlue, UpperBlue)#in range then set to 1, 白色； 0=背景黑色
    Things= cv.bitwise_and(image_orig,image_orig,mask=mask)#图像与运算，类似于显示原图中的区域
    image_Segment_Blue=Things
    return image_Segment_Blue


def Segment_Yellow(image_orig,image_HSV,H,S,V):#must be HSV 
    logging.info('Segment_Yellow...')
    H_Yellow_Min=15
    H_Yellow_Max=26
    S_Yellow_Min=43
    S_Yellow_Max=255
    V_Yellow_Min=46
    V_Yellow_Max=255
    LowerYellow = np.array([H_Yellow_Min, S_Yellow_Min,V_Yellow_Min])
    UpperYellow = np.array([H_Yellow_Max, S_Yellow_Max,V_Yellow_Max])
    mask= cv.inRange(image_HSV, LowerYellow, UpperYellow)#in range then set to 1, 白色； 0=背景黑色
    Things= cv.bitwise_and(image_orig,image_orig,mask=mask)#图像与运算，类似于显示原图中的区域
    image_Segment_Yellow=Things
    
    return image_Segment_Yellow
    

#计算边界/轮廓
#func explaination https://blog.csdn.net/hjxu2016/article/details/77833336
def Edge(path):#should input binary image
    logging.info('Edge...')
    #calculation
    image=cv.imread(path,cv.CV_8UC1) 
    binary,contours, hierarchy = cv.findContours(image,cv.RETR_EXTERNAL,cv.CHAIN_APPROX_NONE) 
    #in open CV3， findContours return 3 value; so my version is openCV3

    #draw boundary on the image
    #image=cv.drawContours(image,contours,-1,(0,0,255),3)#-1:draw all edge;  3:line width；（0，0，255):RGB color scaler
    #image_Edge=image
    return contours #outpur contours is a array, 

#不相连的轮廓求最小外界矩形,并添加到原图中
def draw_Min_Rect(image,contours):
    logging.info('draw_Min_Rect...')
    N= np.array(contours).shape[0]
    for i in range (0,N):
        min_=cv.minAreaRect(contours[i])
        box = cv.boxPoints(min_) 
        box = np.int0(box)
        cv.drawContours(image, [box], 0, (0,0,255), 3)
    #cv.imshow('0', image)
    #cv.waitKey(0)# the number of windows
    image_with_MinRect=image
    return image_with_MinRect


#Area筛选box
def Rect_filter(contours):
    logging.info('Rect_filter...')
    N= np.array(contours).shape[0]
    contours_update=[]
    Box_update=[]
    for i in range (0,N):
        Rect=cv.minAreaRect(contours[i])
        Area_=Rect[1][0]*Rect[1][1]
        Box=cv.boxPoints(Rect)
        if int(Area_) >=1000 and int(Area_)<=15000 : 
            logging.info('W:%s, H:%s, Ratio:%s'%(Rect[1][0], Rect[1][1],(Rect[1][0])/(Rect[1][1])))
            logging.info( 'Area:%s'%Area_)
            if abs(Rect[2])<=10:
                if (Rect[1][0])/(Rect[1][1])>=2 and (Rect[1][0])/(Rect[1][1])<=5:   
                    contours_update.append(contours[i])
                    Box_update.append(Box)
            else:
                if (Rect[1][1])/(Rect[1][0])>=2 and (Rect[1][1])/(Rect[1][0])<=5:   
                    contours_update.append(contours[i])
                    Box_update.append(Box)
    return contours_update,Box_update

def Angle_filter(conts):
    logging.info('Angle_filter...')
    N= np.array(conts).shape[0]
    Box_update=[]
    contours_update=[]
    for i in range (0,N):
        Rect=cv.minAreaRect(conts[i])
        A=Rect[2]#anglle\
        Box=cv.boxPoints(Rect) 
        if abs(A)>=80 and abs(A)<=90:
            contours_update.append(conts[i])
            Box_update.append(Box)
        else:
            if abs(A)<=10:
                contours_update.append(conts[i])
                Box_update.append(Box)
    return contours_update,Box_update

###############################Sobel_seg func additional

#高斯模糊，降噪
def Gauss(image):
    logging.info('Gauss...')
    image_Guass=cv.GaussianBlur(image, (5, 5), 1.5)
    return image_Guass

#图像增强/边缘检测，sobel算子（选定水平边缘为检测对象）
def Enhencement(image):
    logging.info('Enhencement...')

    Sobel_x=0 # x horizontal，
    Sobel_y= cv.Sobel(image, cv.CV_16S,dx=1,dy=0,ksize=3)
    #dy=1 means vertical, ddepth=CV_16S;,ksize use default;ksize must be 1,3,5,7, what is default?=3
    image_Sobel=Sobel_y
    return image_Sobel

#二值化， 两种方式：全局阈值和 自适应阈值cv.adaptiveThreshold
def Binary(image):
    logging.info('Binary...')
    # for binary type, donot need :MaxvalMaxval=conf.getint('Binary', 'value')# set to be that value, 
    return_value,image_Binary=cv.threshold(image,127,255,cv.THRESH_BINARY)
    return image_Binary

#闭操作，区域连接
def Close(image):# should input binary image
    logging.info('Close...')
    kernel=cv.getStructuringElement(cv.MORPH_RECT,(22,22))  #must be int 方阵
    #func explaination   https://blog.csdn.net/qq_31186123/article/details/78770141
    image_Close=cv.morphologyEx(image,cv.MORPH_CLOSE,kernel)

    return image_Close


#######################Cut_image
#截取车牌box，
def CutImage(path_,Box):#Box_update is minAreaRect
    logging.info('CutImage...')
    image_orig=cv.imread(path_)
    points=np.int0(Box)
    shift=4
    left_min=min(points[0][0],points[1][0],points[2][0],points[3][0])
    right_max=max(points[0][0],points[1][0],points[2][0],points[3][0])
    up_max=max(points[0][1],points[1][1],points[2][1],points[3][1])
    down_min=min(points[0][1],points[1][1],points[2][1],points[3][1])
    CutImage=image_orig[int(down_min-shift):int(up_max+shift),int(left_min-shift):int(right_max+shift)]

    return CutImage

#统一尺寸，classical 136*36， 但是LeNet 是默认36*36
def Unisize(CutImage):
    logging.info('Unisize...')
    width=136
    hight=36
    image_Unisize=cv.resize(CutImage,(width,hight))
    
    return image_Unisize

#######################processing func
#1）
def Color_HSV_contours(image_orig,path_save_color,color):
    image_Equalized=Equalization(image_orig,path_save_color)
    save_image(image_Equalized,'image_Equalized',path_save_color)
    
    image_HSV,H,S,V=HSV(image_Equalized,path_save_color)
    save_image(image_HSV,'image_HSV',path_save_color)
    ##print np.array(image_HSV).shape #is h, [1] is w, [2] is ch
    
    ###
    if color=='Blue':
        image_Segment_Blue=Segment_Blue(image_orig,image_HSV,H,S,V)
        save_image(image_Segment_Blue,'image_Segment_Blue',path_save_color) 
    
        contours=Edge(path_save_color+'image_Segment_Blue.jpg')
        #os.remove(path_save_color+'image_Segment_Blue.jpg')
    else:
        if color=='Yellow':
            image_Segment_Yellow=Segment_Yellow(image_orig,image_HSV,H,S,V)
            save_image(image_Segment_Yellow,'image_Segment_Yellow',path_save_color) 
            contours=Edge(path_save_color+'image_Segment_Yellow.jpg')
            #os.remove(path+'image_Segment_Yellow.jpg')
    
    return contours
    


def Color_seg(image_orig,path_save_color):
    logging.info('\n###################################Color_seg begining...')
    
    contours=Color_HSV_contours(image_orig,path_save_color,'Blue')#
    #filter
    contours_update,Box_update=Angle_filter(contours)
    contours_update,Box_update=Rect_filter(contours_update)
    #print np.array(contours_update).shape[0]

    #Blue
    if np.array(contours_update).shape[0]!=0: #Blue if 0 or >3 then the car and plate have same color  
        image_with_MinRect_update=draw_Min_Rect(image_orig,contours_update)# draw on the image_orig object  
        save_image(image_with_MinRect_update,'image_with_MinRect_update',path_save_color)

    else: # Yellow
        logging.info('！！Blue failed, try Yellow...')
        contours=Color_HSV_contours(image_orig,path_save_color,'Yellow')#
        #filter
        contours_update,Box_update=Angle_filter(contours)
        contours_update,Box_update=Rect_filter(contours_update)
        #print np.array(contours_update).shape[0]
        
        image_with_MinRect_update=draw_Min_Rect(image_orig,contours_update)# draw on the image_orig object  
        save_image(image_with_MinRect_update,'image_with_MinRect_update',path_save_color)

    logging.info('The final box of Color_seg:%s'%np.array(contours_update).shape[0])
    return contours_update,Box_update
    
      
                
    

##2）
def Sobel_contours(image_orig,path_save_sobel):
    image_Gauss=Gauss(image_orig)
    save_image(image_Gauss,"image_Gauss",path_save_sobel)
    
    iamge_Gary=Gary(image_Gauss)
    save_image(iamge_Gary,"image_Gary",path_save_sobel)
    
    image_Sobel=Enhencement(iamge_Gary)
    save_image(image_Sobel,"image_Sobel",path_save_sobel)
    
    image_Binary=Binary(image_Sobel)
    save_image(image_Binary,"image_Binary",path_save_sobel)
    
    image_Close=Close(image_Binary)
    save_image(image_Close,"image_Close",path_save_sobel)
    
    image_Close=Close(image_Close)
    save_image(image_Close,"image_Close",path_save_sobel)
    
    path_edge=path_save_sobel+'image_Close.jpg'
    contours=Edge(path_edge)
    
    return contours
    

def Sobel_seg(image_orig,path_save_sobel): 
    logging.info('\n###################################Sobel_seg begining...')
    
    contours=Sobel_contours(image_orig,path_save_sobel)
    
    contours_update,Box_update=Angle_filter(contours)
    contours_update,Box_update=Rect_filter(contours_update)
    image_with_MinRect_update=draw_Min_Rect(image_orig,contours_update)
    save_image(image_with_MinRect_update,"image_with_MinRect_update",path_save_sobel)

    logging.info('The final box of Sobel_seg:%s'%np.array(contours_update).shape[0])
    
    return contours_update,Box_update
                     
    
    

##3）Segment
def Classify(path_,Box_update):
    logging.info('Classify...')
    for i in Box_update:
        image_Cut=CutImage(path_,i)
        image_Unisize=Unisize(image_Cut)
        
        save_image(image_Unisize,'image_Unisize',path_tmp)
        
        X_=[]
        #classify
        img=cv.imread(path_tmp+'image_Unisize.jpg',cv.IMREAD_GRAYSCALE)
        N=(np.array(img).shape[1])*(np.array(img).shape[0])
        img=np.reshape(np.array(img),N)
        X_.append(img.tolist())
        
        Y_predict=svm.predict(np.array(X_,dtype=np.float32))
        result=int(Y_predict[1][0][0])
        if int(result)==1:
            image_located=image_Unisize
            save_image(image_located,'plate',path_plate)
            logging.info('plate located, break the loop')
            break
    
        logging.info('SVM result: There is no plate in images that located!')    
        
        
        
#########################################
#########################################

if __name__ == '__main__':
    ini()
    path='./image_preprocess/'
    filelist=os.listdir(path)
    #file_=args["image_name"]
    file_='orig7.jpg'
    path_=path+file_
    path_tmp='./image_preprocess/tmp/'
    path_plate='./image_preprocess/plate/'
    path_save_color='./image_preprocess/Color_seg/'
    path_save_sobel='./image_preprocess/Sobel_seg/'
    clean_folder(path_save_color)
    clean_folder(path_save_sobel)
    #model=args["model_type"] #0:normal,1:raw_blue,2:raw_yellow,  other value :raw_sobel
    model=0
    try:
        if file_ not in filelist:
            raise ValueError('ImageNameError')
        else:
            image_orig=cv.imread(path_)
    except ValueError:
        logging.info('File not found!',exc_info=True)
        raise
        
    try:    
        svm=cv.ml.SVM_load('./model/mdoel.m')
    except:
        logging.info('ModelPathError, svm load model failed')
    
    if model==0:
        logging.info('##filter model')
        contours_update,Box_update=Color_seg(image_orig,path_save_color)
        if np.array(contours_update).shape[0]==0:
            logging.info('！！Color_locate failed, try Sobel_locate...')
            contours_update,Box_update=Sobel_seg(image_orig,path_save_sobel)
    else:
        if model==1:
            logging.info('##raw blue model')
            contours=Color_HSV_contours(image_orig,path_save_color,'Blue')
            image_with_MinRect=draw_Min_Rect(image_orig,contours)
            save_image(image_with_MinRect,'image_with_MinRect',path_save_color) 
        else:
            if model==2:
                logging.info('##raw yellow model')
                contours=Color_HSV_contours(image_orig,path_save_color,'Yellow')
                image_with_MinRect=draw_Min_Rect(image_orig,contours)
                save_image(image_with_MinRect,'image_with_MinRect',path_save_color) 

            else:
                logging.info('##raw sobel model')
                contours=Sobel_contours(image_orig,path_save_sobel)
                image_with_MinRect=draw_Min_Rect(image_orig,contours)
                save_image(image_with_MinRect,"image_with_MinRect",path_save_sobel)        
    
    Classify(path_,Box_update)

```
